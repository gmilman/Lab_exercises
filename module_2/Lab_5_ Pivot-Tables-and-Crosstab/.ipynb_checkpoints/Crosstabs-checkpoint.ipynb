### 1. Import the moovies dataset 

# https://www.kaggle.com/carolzhangdc/imdb-5000-movie-dataset

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('movie_metadata.csv')
df
```

### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(index=df['country'],columns='count', values= df['country'], aggfunc=np.size)
```

```python
pd.crosstab(df['country'],columns='count', values= df['country'], aggfunc=np.size)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab(df.director_name, columns='count', values =df['director_name'], aggfunc=np.size).sort_values('count', ascending=False)
```

### 4. Average duration of filmes grouped by color

```python
list(df.columns)
```

```python
pd.crosstab(df.color,columns= 'average_duration',  values=df.duration, aggfunc=np.mean)
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
# lets calculate likes per film in the df. director like + actor likes)
df['total_likes']= df.director_facebook_likes + df.actor_1_facebook_likes + df.actor_2_facebook_likes + df.actor_3_facebook_likes
pd.set_option('display.max_columns', None)
df
```
