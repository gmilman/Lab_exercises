#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending'
````

````python
# your code here
page=requests.get(url)
soup=BeautifulSoup(page.content, 'lxml')
print(soup.prettify())
````

````python
repo_names=[ repo.text.strip() for repo in soup.find_all('h1', class_='h3 lh-condensed')]

repo_names=[repo.replace('\n','') for repo in repo_names]
repo_names=[repo.replace(' /      ', ' / ') for repo in repo_names]
repo_names
````

#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
# your code here
page=requests.get(url)
soup=BeautifulSoup(page.content, 'lxml')
image_links= soup.find_all('a', class_='image')
image_links

image_info=[]
for link in image_links:
    image_tag = link.findChildren("img")
    image_info.append((image_tag[0]["src"], image_tag[0]["alt"]))
# add https to beginning of str
image_info
````
