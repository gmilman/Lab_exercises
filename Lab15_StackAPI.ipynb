 Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 
        no question posted today so went back to beginning of month

```python
from datetime import datetime, timedelta

today =datetime(2023,1,26,0,0,0)

SITE.page_size = 10 ## limts the number of answers per page 
SITE.max_pages = 1 ## limits the number of pages

questions_5_star = SITE.fetch('questions', fromdate=today, min=5, tagged = 'python', sort='votes')
pprint(questions_5_star)
```
